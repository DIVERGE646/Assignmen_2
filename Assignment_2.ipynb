```python
2.Explain the key features of lists in Python

ANSWER:

Lists  are mutable, ordered collections that can store a variety of object types.

1. Ordered
Lists maintain the order of elements, meaning the order in which elements are
inserted is preserved. Each element in the list has an index, starting from 0.
my_list = [10, 20, 30]
print(my_list[0])  # Output: 10
print(my_list[1])  # Output: 20

2. Mutable
Lists are mutable, meaning you can modify their elements after creation.
one can add, remove, or change items within a list.
my_list = [1, 2, 3]
my_list[1] = 20
print(my_list)  # Output: [1, 20, 3]

3. Dynamic
Lists can grow and shrink dynamically in size.One can append, insert,
or remove elements without declaring a fixed size.
my_list = [1, 2, 3]
my_list.append(4)  # Adding an element
print(my_list)  # Output: [1, 2, 3, 4]

4. Heterogeneous Elements
Lists can store elements of different data types, such as integers, strings,
floats, or even other lists.
my_list = [1, "hello", 3.14, [10, 20]]
print(my_list)  # Output: [1, 'hello', 3.14, [10, 20]]

5. Indexed Access
one can access elements in a list using their index. Positive indices start
from 0, while negative indices allow access from the end of the list.
my_list = [10, 20, 30]
print(my_list[0])   # Output: 10 (first element)
print(my_list[-1])  # Output: 30 (last element)

6. Supports Slicing
Lists support slicing, which allows one to extract a sub-list by specifying
a range of indices.
my_list = [1, 2, 3, 4, 5]
print(my_list[1:4])  # Output: [2, 3, 4]

7. Iterable
Lists are iterable loop through the elements using a for
loop or other iteration methods.
my_list = [1, 2, 3]
for item in my_list:
    print(item)
# Output:
# 1
# 2
# 3
8. Supports List Comprehensions
List comprehensions provide a concise way to create lists based on existing
iterables or conditions.
squares = [x**2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]

9. Commonly Used List Methods
append(): Adds an element to the end of the list.
my_list.append(6)

extend(): Adds multiple elements to the list.
my_list.extend([7, 8])

insert(): Inserts an element at a specified position.
my_list.insert(1, "new")

remove(): Removes the first occurrence of an element.
my_list.remove(2)

pop(): Removes and returns an element at a given index.
my_list.pop(2)

sort(): Sorts the list in place.
my_list.sort()

reverse(): Reverses the elements of the list in place.
my_list.reverse()
```

```python
4.Compare and contrast tuples and lists with examples

ANSWER:



Both tuples and lists are sequence data types in  that can store collections of
items. However, they differ in several key characteristics, making each suitable
 for different use cases. Below is a comparison of tuples and lists with examples:
________________________________________
1. Mutability
•	Lists: Mutable – meaning the elements of a list can be changed (added, removed,
or modified) after the list is created.
•	Tuples: Immutable – once a tuple is created, its elements cannot be changed,
added, or removed.
Example:


# List (Mutable)
my_list = [1, 2, 3]
my_list[1] = 20  # Modifying the second element
print(my_list)  # Output: [1, 20, 3]

# Tuple (Immutable)
my_tuple = (1, 2, 3)
# my_tuple[1] = 20  # Error! Tuples do not support item assignment
________________________________________
2. Syntax
•	Lists: Defined using square brackets [ ].
•	Tuples: Defined using parentheses ( ).
Example:


my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

4. Usability
•	Lists: Used when one need a collection of elements that may need to change
during the execution of a program.
•	Tuples: Used when the collection of elements is fixed or should not change,
such as storing constants or multiple return values from a function.
Example:


# List Use Case
shopping_list = ["apples", "bananas", "grapes"]
shopping_list.append("oranges")  # One can add items to the list
print(shopping_list)  # Output: ['apples', 'bananas', 'grapes', 'oranges']

# Tuple Use Case
coordinates = (52.5163, 13.3777)  # Coordinates don't change
________________________________________
5. Functions and Methods
•	Lists: Have many built-in methods for modifying their contents such as
append(), remove(), extend(), pop(), and sort().
•	Tuples: Have fewer methods because they cannot be modified. They only have
count() and index().
Example:


# List methods
my_list = [10, 20, 30]
my_list.append(40)
print(my_list)  # Output: [10, 20, 30, 40]

# Tuple methods
my_tuple = (10, 20, 30, 10)
print(my_tuple.count(10))  # Output: 2 (counts the occurrence of 10)
print(my_tuple.index(20))  # Output: 1 (gives the index of the first occurrence of 20)
________________________________________
6. Size and Memory Usage
•	Lists: Due to their mutable nature, lists generally consume more memory than tuples.
•	Tuples: As immutable structures, tuples are more memory-efficient than lists.
Example:


import sys

my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print(f"List size: {sys.getsizeof(my_list)} bytes")  # Output: List size in bytes
print(f"Tuple size: {sys.getsizeof(my_tuple)} bytes")  # Output: Tuple size in bytes
________________________________________
7. Immutability and Hashing
•	Lists: Cannot be used as dictionary keys or set elements because they are mutable and unhashable.
•	Tuples: Can be used as dictionary keys or set elements because they are immutable and hashable.
Example:


# Using a tuple as a dictionary key
my_dict = { (1, 2): "coordinates", (3, 4): "more coordinates" }
print(my_dict[(1, 2)])  # Output: 'coordinates'

# Using a list as a dictionary key would result in an error
# my_dict = { [1, 2]: "coordinates" }  # Error: TypeError: unhashable type: 'list'

```

```python
5.Describe the key features of sets and provide examples of their use

ANSWER:

Key Features of Sets in
A set in  is an unordered collection of unique elements. It is a versatile data
structure that supports various operations, especially when one need to handle
collections of data without duplicates or when one need fast membership testing.
________________________________________
Key Features of Sets:
1. Unordered
Sets do not maintain any order. The elements in a set do not have a defined order,
 and one cannot rely on the insertion order when iterating over a set.
Example:
my_set = {3, 1, 2}
print(my_set)  # Output might be: {1, 2, 3} or {3, 2, 1}, order is arbitrary

2. Unique Elements
Sets automatically remove duplicates. If one add duplicate values to a set, only
 one instance will be stored.
Example:
my_set = {1, 2, 2, 3, 4, 4}
print(my_set)  # Output: {1, 2, 3, 4}

3. Mutable
While the elements of a set must be immutable (e.g., numbers, strings, tuples),
the set itself is mutable, meaning one can add or remove elements after the set is created.
Example:
my_set = {1, 2, 3}
my_set.add(4)  # Adding an element
print(my_set)  # Output: {1, 2, 3, 4}

4. No Indexing or Slicing
Since sets are unordered, they do not support indexing or slicing like lists or
tuples. One cannot access elements by position.
Example:
my_set = {10, 20, 30}
# print(my_set[0])  # Error: 'set' object is not subscriptable

5. Efficient Membership Testing
Sets provide fast membership testing because they are implemented using hash
tables. This makes operations like checking whether an element is in the set
 (in operator) very efficient, typically with O(1) time complexity.
Example:
my_set = {1, 2, 3}
print(2 in my_set)  # Output: True
print(5 in my_set)  # Output: False

6. Set Operations (Union, Intersection, Difference, Symmetric Difference)
Sets support various mathematical operations like union, intersection, difference,
and symmetric difference. These operations are useful for working with collections
of data.
•	Union (| or union()): Combines all unique elements from both sets.
•	Intersection (& or intersection()): Returns elements common to both sets.
•	Difference (- or difference()): Returns elements present in one set but not
in the other.
•	Symmetric Difference (^ or symmetric_difference()): Returns elements that are
in either of the sets but not in both.
Example:
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

print(set_a | set_b)  # Union: {1, 2, 3, 4, 5, 6}
print(set_a & set_b)  # Intersection: {3, 4}
print(set_a - set_b)  # Difference: {1, 2}
print(set_a ^ set_b)  # Symmetric Difference: {1, 2, 5, 6}

7. Immutable Sets (frozenset)
 also has an immutable version of sets called frozenset. Once created, the
 elements of a frozenset cannot be modified (similar to a tuple). frozenset
 supports all the same set operations, but the set itself cannot be changed.
Example:
my_frozenset = frozenset([1, 2, 3])
# my_frozenset.add(4)  # Error: 'frozenset' object has no attribute 'add'

8. Duplicate Removal
Sets are commonly used to remove duplicates from a list or other collection.
Example:
my_list = [1, 2, 2, 3, 4, 4, 5]
unique_elements = set(my_list)  # Converts the list to a set, removing duplicates
print(unique_elements)  # Output: {1, 2, 3, 4, 5}
9. Modifying Sets
One can add or remove elements from a set using the following methods:
•	add(): Adds a single element to the set.
•	remove(): Removes a specific element from the set. Raises an error if the
element is not present.
•	discard(): Removes an element from the set without raising an error if the
element is not present.
•	pop(): Removes and returns an arbitrary element from the set
 (since sets are unordered).
•	clear(): Removes all elements from the set.
Example:
my_set = {1, 2, 3}
my_set.add(4)  # Adds 4 to the set
my_set.remove(2)  # Removes 2 from the set
my_set.discard(5)  # Does nothing since 5 is not in the set (no error)
my_set.pop()  # Removes and returns an arbitrary element
print(my_set)  # Output: remaining elements in the set

```

```python
 7.Describe how to add, modify, and delete items in a dictionary with examples

 ANSWER:

 Adding, Modifying, and Deleting Items in a Dictionary in
A dictionary in  is a collection of key-value pairs where each key must be unique
and immutable, while the values can be of any type. Dictionaries are mutable,
allowing one to add, modify, and delete key-value pairs easily.
________________________________________
1. Adding Items to a Dictionary
One can add new key-value pairs to a dictionary by simply assigning a value to a new key.
Example:


# Creating an empty dictionary
my_dict = {}

# Adding a new key-value pair
my_dict["name"] = "Sujan"
my_dict["age"] = 25

print(my_dict)  # Output: {'name': 'Sujan', 'age': 25}
In the example above, we added a new key "name" with the value "Sujan", and a key
"age" with the value 25.
________________________________________
2. Modifying Items in a Dictionary
To modify the value associated with an existing key, simply assign a new value to that key. If the key already exists, the old value will be replaced with the new value.
Example:


# Existing dictionary
my_dict = {"name": "Sujan", "age": 25, "city": "New York"}

# Modifying the value of the key "age"
my_dict["age"] = 26

# Modifying the value of the key "city"
my_dict["city"] = "Los Angeles"

print(my_dict)  # Output: {'name': 'Sujan', 'age': 26, 'city': 'Los Angeles'}
In the above example, the value of "age" was changed from 25 to 26, and the value
of "city" was updated from "New York" to "Los Angeles".
________________________________________
3. Deleting Items from a Dictionary
There are several ways to delete items (key-value pairs) from a dictionary:
a. Using del Statement
One can use the del statement to remove a specific key-value pair by specifying the key.
Example:


my_dict = {"name": "Sujan", "age": 26, "city": "Los Angeles"}

# Deleting the key "age"
del my_dict["age"]

print(my_dict)  # Output: {'name': 'Sujan', 'city': 'Los Angeles'}
b. Using pop() Method
The pop() method removes the key and returns its associated value. If the key
doesn’t exist, it will raise a KeyError unless one provide a default value.
Example:


my_dict = {"name": "Sujan", "age": 26, "city": "Los Angeles"}

# Removing the key "city" and returning its value
removed_value = my_dict.pop("city")
print(removed_value)  # Output: Los Angeles
print(my_dict)  # Output: {'name': 'Sujan', 'age': 26}
c. Using popitem() Method
The popitem() method removes and returns the last inserted key-value pair in the
 dictionary. In  3.7+, dictionaries maintain the insertion order, so it removes
 the most recently added item.
Example:


my_dict = {"name": "Sujan", "age": 26, "city": "Los Angeles"}

# Removing the last inserted key-value pair
last_item = my_dict.popitem()
print(last_item)  # Output: ('city', 'Los Angeles')
print(my_dict)  # Output: {'name': 'Sujan', 'age': 26}
d. Using clear() Method
The clear() method removes all items from the dictionary, leaving it empty.
Example:


my_dict = {"name": "Sujan", "age": 26, "city": "Los Angeles"}

# Clearing the dictionary
my_dict.clear()

print(my_dict)  # Output: {}

```

```python
8.Discuss the importance of dictionary keys being immutable and provide examples.

#ANSWER:

In , dictionary keys must be immutable because dictionaries use a hashing mechanism
 to map keys to their associated values. The key's immutability ensures that the
 hash value of a key remains constant over its lifetime, allowing for efficient
 lookups, insertions, and deletions. If keys were mutable, their hash values could change,
  making it impossible to retrieve the correct values associated with those keys.
________________________________________
Key Reasons for Immutability:
1. Hashing and Efficient Lookups
•	Dictionaries in  are implemented as hash tables. When a key-value pair is added,
 computes a hash of the key, which determines where the pair will be stored.
•	Immutability guarantees that once the hash of a key is computed, it remains the
same throughout the key's existence. This ensures that the dictionary can reliably
find the value associated with a key.
•	Mutable objects (like lists) can change, which would alter the hash value, making
it difficult (or impossible) to retrieve the associated value using the modified key.
Example:


my_dict = {1: "one", "name": "Alice", (10, 20): "coordinates"}
print(my_dict[1])  # Output: "one"
print(my_dict[(10, 20)])  # Output: "coordinates"
Here, keys like 1, "name", and the tuple (10, 20) are immutable, meaning their
hash value remains consistent, allowing the dictionary to retrieve the values efficiently.
2. Ensuring Consistency of Key-Value Mappings
•	Dictionary lookups rely on the key remaining constant. If a key could change
 (e.g., if it were a list), it would lead to unpredictable behavior.
•	By enforcing that keys must be immutable,  ensures that each key's position
in the hash table is stable, and the dictionary can always locate the corresponding value.
Example of an Issue with Mutable Keys: If  allowed mutable keys like lists,
the following scenario would cause problems:


# Hypothetical (not allowed in )
mutable_key = [1, 2, 3]
my_dict = {mutable_key: "value"}

# Changing the mutable key
mutable_key.append(4)

# The dictionary would now fail to locate the key
print(my_dict[mutable_key])  # This would raise a KeyError because the hash has changed.
 avoids such issues by not allowing mutable types (e.g., lists, sets) as dictionary keys.

```
